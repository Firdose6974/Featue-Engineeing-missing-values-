# count or Frequency encoding

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

pd.pandas.set_option("display.max_columns",None)
pd.pandas.set_option("display.max_rows",None)
```

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header = None,index_col=None)
train_set.head()
```

```python
train_set[1].value_counts().index[0]
```

```python
# Tkaing out the categorical features from the dataset
Categorical_features = [feature for feature in train_set.columns if train_set[feature].dtypes =="O"]
train_set[Categorical_features].head()
```

```python
# another method for taking out the categorical features from the dataset
columns = [1,3,5,6,7,8,9,13,14]
```

```python
train_set[columns].head()
```

```python
test_set = train_set[columns]
test_set.columns = ["Employment", "Degree", "Status", "Designation", "Family_job","Race", "Sex", "Country", "Income"]
```

```python
test_set.columns
```

```python
test_set.head()
```

```python
for feature in test_set.columns:
    print(feature, ":", len(test_set[feature].unique()), "labels")
```

```python
test_set["Country"].value_counts()
```

```python
Country_map = test_set["Country"].value_counts().to_dict()
```

```python
test_set["Country"]= test_set["Country"].map(Country_map)    # replacing the country column with respect to value_count of each category in country
```

```python
test_set.head()
```

```python
test_set.head(30)
```

# Advantages
- easy to use
- Not increasing feature space

# Disadvantages

- It will provide same weight if the frequencies are same(If 2 countries has same count then it will not handle that type of situation)
